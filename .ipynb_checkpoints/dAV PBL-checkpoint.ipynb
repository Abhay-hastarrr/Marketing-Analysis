```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Data Cleaning

Handle Missing Values



```python
# Drop rows with missing values in critical columns
df.dropna(subset=['Income', 'Year_Birth'], inplace=True)

# Impute missing values for other columns (e.g., median for numerical, mode for categorical)
df['Education'].fillna(df['Education'].mode()[0], inplace=True)
df['Marital_Status'].fillna(df['Marital_Status'].mode()[0], inplace=True)

print("\nMissing Values After Cleaning:")
print(df.isnull().sum())
```

Remove Duplicates

```python
df.drop_duplicates(inplace=True)
print(f"\nDataset Size After Removing Duplicates: {df.shape}")
```

Correct Data Types

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
print("\nData Types After Correction:")
print(df.dtypes)
```

# Remove Outliers

```python
from scipy.stats import zscore

# Select numerical columns
numerical_cols = ['Income', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 
                  'MntSweetProducts', 'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases', 
                  'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth']

# Calculate Z-scores and create new columns
zscore_columns = [col + '_ZScore' for col in numerical_cols]  # Create new column names
df[zscore_columns] = df[numerical_cols].apply(zscore)  # Apply Z-score calculation

# Define threshold (e.g., Z-score > 3 or < -3)
threshold = 3

# Remove outliers
df_cleaned = df[(abs(df[zscore_columns]) <= threshold).all(axis=1)]

print(f"\nDataset Size After Removing Outliers: {df_cleaned.shape}")
```

```python
# Boxplot for Income before and after outlier removal
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x=df['Income'])
plt.title('Income Distribution (Before Outlier removal)')

plt.subplot(1, 2, 2)
sns.boxplot(x=df_cleaned['Income'])
plt.title('Income Distribution (After Outlier Removal)')

plt.tight_layout()
plt.show()
```

Correlation Matrix

```python
# Define the magnify function for hover effects
def magnify():
    return [
        dict(selector="th", props=[("font-size", "7pt")]),
        dict(selector="td", props=[('padding', "0em 0em")]),
        dict(selector="th:hover", props=[("font-size", "12pt")]),
        dict(selector="tr:hover td:hover", props=[('max-width', '200px'), ('font-size', '12pt')])
    ]

# Function to generate the styled correlation matrix
def corr_matrix(df):
    # Compute the correlation matrix
    corr = df.corr()
    
    # Create a styled DataFrame with enhanced formatting
    vis = (
        corr.style
        .background_gradient(cmap='coolwarm', axis=1)  # Gradient background
        .set_properties(**{'max-width': '80px', 'font-size': '10pt'})  # Default cell properties
        .set_caption("Hover to magnify")  # Table caption
        .format(precision=2)  # Format numbers to 2 decimal places
        .set_table_styles(magnify())  # Apply hover effects
    )
    
    return vis
    
# Filter only numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Generate and display the styled correlation matrix
styled_corr_matrix = corr_matrix(numeric_df)
styled_corr_matrix
```

# Feature Engineering


One-Hot Encoding for Categorical Variables

```python
df_encoded = pd.get_dummies(df, columns=['Education', 'Marital_Status'], drop_first=True)
print("\nEncoded Dataset:")
df_encoded.head()
```

Create New Features

```python
df_encoded['Engagement_Score'] = (
    df_encoded['AcceptedCmp1'] + 
    df_encoded['AcceptedCmp2'] + 
    df_encoded['AcceptedCmp3'] + 
    df_encoded['AcceptedCmp4'] + 
    df_encoded['AcceptedCmp5']
) 
print("\nDataset with New Features:")
df_encoded.head()
```

# Task 1: Which campaigns have the highest engagement rates

```python
# Calculate engagement rates for each campaign
campaign_columns = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']
engagement_rates = df_encoded[campaign_columns].mean() * 100

# Display results
print("Engagement Rates for Each Campaign:")
print(engagement_rates)

# Visualize engagement rates
plt.figure(figsize=(10, 6))
sns.barplot(x=engagement_rates.index, y=engagement_rates.values)
plt.title('Engagement Rates Across Campaigns')
plt.ylabel('Engagement Rate (%)')
plt.xlabel('Campaign')
plt.show()
```

# Task 2: How does customer demographic influence conversion rates

```python
# Calculate conversion rates by demographic groups
df_encoded['Age'] = 2023 - df_encoded['Year_Birth']

# Example: Conversion rate by Age group
bins = [0, 30, 40, 50, 60, 100]
labels = ['<30', '30-40', '40-50', '50-60', '>60']
df_encoded['Age_Group'] = pd.cut(df_encoded['Age'], bins=bins, labels=labels)

conversion_by_age = df_encoded.groupby('Age_Group')['Response'].mean() * 100
print("Conversion Rates by Age Group:")
print(conversion_by_age)

# Visualize conversion rates by Age Group
plt.figure(figsize=(10, 6))
sns.barplot(x=conversion_by_age.index, y=conversion_by_age.values)
plt.title('Conversion Rates by Age Group')
plt.ylabel('Conversion Rate (%)')
plt.xlabel('Age Group')
plt.show()
```

# Task 3: Which channels are most effective

```python
avg_web_visits = df['NumWebVisitsMonth'].mean()
avg_web_purchases = df['NumWebPurchases'].mean()
avg_catalog_purchases = df['NumCatalogPurchases'].mean()
avg_store_purchases = df['NumStorePurchases'].mean()

# Calculate conversion rates
conversion_rate_web = (avg_web_purchases / avg_web_visits) * 100
conversion_rate_catalog = (avg_catalog_purchases / avg_web_visits) * 100
conversion_rate_store = (avg_store_purchases / avg_web_visits) * 100

# Print conversion rates
print(f"Web Conversion Rate: {conversion_rate_web:.2f}%")
print(f"Catalog Conversion Rate: {conversion_rate_catalog:.2f}%")
print(f"Store Conversion Rate: {conversion_rate_store:.2f}%")

# Create a DataFrame for visualization
conversion_rates = pd.DataFrame({
    'Channel': ['Web', 'Catalog', 'Store'],
    'Conversion Rate (%)': [conversion_rate_web, conversion_rate_catalog, conversion_rate_store]
})

# Bar plot for conversion rates
plt.figure(figsize=(10, 6))
sns.barplot(data=conversion_rates, x='Channel', y='Conversion Rate (%)', palette='viridis')
plt.title('Conversion Rates by Purchase Channel')
plt.ylabel('Conversion Rate (%)')
plt.xlabel('Purchase Channel')
plt.show()

# Pie chart for relative contribution of each channel
plt.figure(figsize=(8, 8))
plt.pie(conversion_rates['Conversion Rate (%)'], labels=conversion_rates['Channel'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette('viridis'))
plt.title('Relative Contribution of Conversion Rates by Channel')
plt.show()
```

# Task 4: Are promotions more successful on specific days or times

```python
# Extract day-of-week and month
df_encoded['DayOfWeek'] = df_encoded['Dt_Customer'].dt.day_name()

# Response rate by Day of Week
response_by_day = df_encoded.groupby('DayOfWeek')['Response'].mean() * 100
print("Response Rates by Day of Week:")
print(response_by_day)

# Visualize
plt.figure(figsize=(10, 6))
sns.barplot(x=response_by_day.index, y=response_by_day.values)
plt.title('Response Rates by Day of Week')
plt.ylabel('Response Rate (%)')
plt.xlabel('Day of Week')

plt.show()
```

# Task 5: How do repeat visitors respond to marketing campaigns

```python
# Define repeat visitors (e.g., those with >1 web visit per month)
df_encoded['Repeat_Visitor'] = df_encoded['NumWebVisitsMonth'] > 1

# Response rate for repeat vs. non-repeat visitors
response_by_repeat = df_encoded.groupby('Repeat_Visitor')['Response'].mean() * 100
print("Response Rates by Repeat Visitor Status:")
print(response_by_repeat)

# Visualize
plt.figure(figsize=(8, 5))
sns.barplot(x=response_by_repeat.index, y=response_by_repeat.values)
plt.title('Response Rates by Repeat Visitor Status')
plt.ylabel('Response Rate (%)')
plt.xlabel('Repeat Visitor')
plt.xticks([0, 1], ['Non-Repeat', 'Repeat'])
plt.show()
```

# Key Task:- Optimize ROI

```python
# Calculate Net Profit per customer
df['Net_Profit'] = df['Z_Revenue'] - df['Z_CostContact']

# Calculate ROI per customer
df['ROI'] = (df['Net_Profit'] / df['Z_CostContact']) * 100

# Display summary statistics
print(df[['Net_Profit', 'ROI']].describe())

# Analyze campaign engagement
campaign_columns = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']
campaign_engagement = df[campaign_columns].sum()
print("Campaign Engagement:")
print(campaign_engagement)

# Analyze purchase channel performance
channel_performance = pd.DataFrame({
    'Channel': ['Web', 'Catalog', 'Store'],
    'Purchases': [
        df['NumWebPurchases'].sum(),
        df['NumCatalogPurchases'].sum(),
        df['NumStorePurchases'].sum()
    ]
})
print("Channel Performance:")
print(channel_performance)

# Visualize campaign engagement
plt.figure(figsize=(10, 6))
sns.barplot(x=campaign_engagement.index, y=campaign_engagement.values, palette='viridis')
plt.title('Campaign Engagement')
plt.xlabel('Campaign')
plt.ylabel('Number of Responses')
plt.show()

# Visualize channel performance
plt.figure(figsize=(10, 6))
sns.barplot(data=channel_performance, x='Channel', y='Purchases', palette='coolwarm')
plt.title('Purchase Channel Performance')
plt.xlabel('Channel')
plt.ylabel('Total Purchases')
plt.show()

# Identify high-value customers
high_value_customers = df[df['Net_Profit'] > df['Net_Profit'].quantile(0.75)]
print("High-Value Customers:")
print(high_value_customers[['Income', 'Kidhome', 'Teenhome', 'Net_Profit', 'ROI']].describe())

# Analyze spending by product category
product_spending = df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum()
print("Product Category Spending:")
print(product_spending)

# Visualize product category spending
plt.figure(figsize=(10, 6))
product_spending.plot(kind='bar', color='skyblue')
plt.title('Spending by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Spending')
plt.xticks(rotation=45)
plt.show()
```

# Model

```python
# Model Training Extensions

# 1. Import necessary libraries
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
from imblearn.over_sampling import SMOTE  # For handling imbalanced classes if needed

# 2. Define target variable and features
# Let's assume we want to predict 'Response' (whether a customer responds to a campaign)
X = df_encoded.drop(['Response', 'Z_CostContact', 'Z_Revenue', 'ID'], axis=1)  # Drop other non-feature columns if needed

# Remove any other columns that shouldn't be features
cols_to_drop = ['Dt_Customer']  # Add any other non-numeric or irrelevant columns
X = X.drop(columns=cols_to_drop, errors='ignore')

# Ensure all data is numeric for modeling
X = X.select_dtypes(include=['int64', 'float64'])

y = df_encoded['Response']

# 3. Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)

# 4. Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 5. Handle class imbalance if necessary
# Check class distribution
print("Class distribution before SMOTE:")
print(y_train.value_counts(normalize=True))

# Apply SMOTE if there's significant imbalance
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_scaled, y_train)

print("Class distribution after SMOTE:")
print(pd.Series(y_train_resampled).value_counts(normalize=True))

# 6. Train multiple models to compare
# Model 1: Logistic Regression
log_reg = LogisticRegression(random_state=42, max_iter=1000)
log_reg.fit(X_train_resampled, y_train_resampled)

# Model 2: Random Forest
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train_resampled, y_train_resampled)

# Model 3: Gradient Boosting
gb = GradientBoostingClassifier(random_state=42)
gb.fit(X_train_resampled, y_train_resampled)

# 7. Evaluate models
models = {
    "Logistic Regression": log_reg,
    "Random Forest": rf,
    "Gradient Boosting": gb
}

for name, model in models.items():
    y_pred = model.predict(X_test_scaled)
    y_pred_proba = model.predict_proba(X_test_scaled)[:,1]
    
    # Print evaluation metrics
    print(f"\nModel: {name}")
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    
    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix - {name}')
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()
    
    # ROC Curve
    fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
    auc = roc_auc_score(y_test, y_pred_proba)
    
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, label=f'AUC = {auc:.3f}')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve - {name}')
    plt.legend(loc='lower right')
    plt.show()

# 8. Hyperparameter tuning for the best model
# Let's assume Random Forest performed best and we want to tune it
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_grid=param_grid,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1
)

grid_search.fit(X_train_resampled, y_train_resampled)

print("\nBest parameters:", grid_search.best_params_)
print("Best ROC AUC score:", grid_search.best_score_)

# 9. Train final model with best parameters
best_rf = grid_search.best_estimator_
best_rf.fit(X_train_resampled, y_train_resampled)

# 10. Evaluate final model
y_pred = best_rf.predict(X_test_scaled)
y_pred_proba = best_rf.predict_proba(X_test_scaled)[:,1]

print("\nFinal Model Evaluation:")
print("Classification Report:")
print(classification_report(y_test, y_pred))

# 11. Feature importance analysis
feature_importances = pd.DataFrame({
    'Feature': X.columns,
    'Importance': best_rf.feature_importances_
}).sort_values('Importance', ascending=False)

print("\nTop 10 Most Important Features:")
print(feature_importances.head(10))

plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=feature_importances.head(15))
plt.title('Feature Importance')
plt.tight_layout()
plt.show()

# 12. Save the model for future use
import joblib
joblib.dump(best_rf, 'marketing_campaign_model.pkl')
joblib.dump(scaler, 'feature_scaler.pkl')

print("\nModel and scaler saved to disk.")
```
